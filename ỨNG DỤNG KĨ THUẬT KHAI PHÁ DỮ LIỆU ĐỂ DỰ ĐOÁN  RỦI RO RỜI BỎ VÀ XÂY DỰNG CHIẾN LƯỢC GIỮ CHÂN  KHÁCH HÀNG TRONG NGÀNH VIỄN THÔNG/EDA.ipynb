# III. Trực quan tỷ lệ rời bỏ của khách hàng theo các biến phân loại

## 1. Biểu đồ thể hiện tỷ lệ rời bỏ theo giới tính

```python
# Tính toán số lượng khách hàng theo giới tính và churn
churn_gender = df.groupby(['gender', 'Churn']).size().unstack(fill_value=0)

# Tính tỷ lệ phần trăm
churn_percent = churn_gender.div(churn_gender.sum(axis=1), axis=0) * 100

# Vẽ stacked bar chart
plt.figure(figsize=(10, 8))
bar1 = plt.bar(churn_gender.index, churn_gender['No'], label='Không rời bỏ', color=sns.color_palette("Blues", 2)[0])
bar2 = plt.bar(churn_gender.index, churn_gender['Yes'], bottom=churn_gender['No'], label='Rời bỏ', color=sns.color_palette("Blues", 2)[1])

# Thêm nhãn số lượng trên cột
for i, gender in enumerate(churn_gender.index):
    total_customers = churn_gender['No'][gender] + churn_gender['Yes'][gender]
    plt.text(i, total_customers + 2, f"{total_customers}", ha='center', fontsize=16, fontweight='bold')

# Thêm nhãn tỷ lệ phần trăm bên trong cột
for i, gender in enumerate(churn_gender.index):
    # Nhãn cho phần "Không rời bỏ"
    plt.text(
        i, 
        churn_gender['No'][gender] / 2,  # Vị trí nhãn ở giữa phần "Không rời bỏ"
        f"{churn_percent['No'][gender]:.1f}%", 
        ha='center', 
        color='black', 
        fontsize=16, 
        fontweight='bold'
    )
    # Nhãn cho phần "Rời bỏ"
    plt.text(
        i, 
        churn_gender['No'][gender] + churn_gender['Yes'][gender] / 2,  # Vị trí nhãn ở giữa phần "Rời bỏ"
        f"{churn_percent['Yes'][gender]:.1f}%", 
        ha='center', 
        color='black', 
        fontsize=16, 
        fontweight='bold'
    )

# Tùy chỉnh biểu đồ
plt.title("Tỷ lệ Churn theo giới tính", fontsize=14)
plt.xlabel("Giới tính")
plt.ylabel("Số lượng khách hàng")
# Customize x-tick labels
plt.xticks(fontsize=14)
plt.tight_layout()
plt.show()
```

## 2. Biểu đồ thể hiện tỷ lệ rời bỏ theo tỷ lệ người cao tuổi

```python
# Tính toán số lượng khách hàng theo SeniorCitizen và Churn
churn_senior = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Tính tỷ lệ phần trăm
churn_percent_senior = churn_senior.div(churn_senior.sum(axis=1), axis=0) * 100

# Vẽ stacked bar chart
plt.figure(figsize=(10, 8))
bar1 = plt.bar(churn_senior.index, churn_senior['No'], label='Không rời bỏ', color=sns.color_palette("Blues", 2)[0])
bar2 = plt.bar(churn_senior.index, churn_senior['Yes'], bottom=churn_senior['No'], label='Rời bỏ', color=sns.color_palette("Blues", 2)[1])

# Thêm nhãn số lượng trên cột
for i, senior in enumerate(churn_senior.index):
    total_customers = churn_senior['No'][senior] + churn_senior['Yes'][senior]
    plt.text(i, total_customers + 2, f"{total_customers}", ha='center', fontsize=16, fontweight='bold')

# Thêm nhãn tỷ lệ phần trăm bên trong cột
for i, senior in enumerate(churn_senior.index):
    # Nhãn cho phần "Không rời bỏ"
    plt.text(
        i, 
        churn_senior['No'][senior] / 2,  # Vị trí nhãn ở giữa phần "Không rời bỏ"
        f"{churn_percent_senior['No'][senior]:.1f}%", 
        ha='center', 
        color='black', 
        fontsize=16, 
        fontweight='bold'
    )
    # Nhãn cho phần "Rời bỏ"
    plt.text(
        i, 
        churn_senior['No'][senior] + churn_senior['Yes'][senior] / 2,  # Vị trí nhãn ở giữa phần "Rời bỏ"
        f"{churn_percent_senior['Yes'][senior]:.1f}%", 
        ha='center', 
        color='black', 
        fontsize=16, 
        fontweight='bold'
    )

# Tùy chỉnh biểu đồ
plt.title("Tỷ lệ Churn theo SeniorCitizen", fontsize=14)
plt.xlabel("SeniorCitizen (0=Không cao tuổi, 1=Cao tuổi)")
plt.ylabel("Số lượng khách hàng")
plt.xticks(ticks=[0, 1], labels=["Không cao tuổi", "Cao tuổi"], fontsize=14)
plt.tight_layout()
plt.show()
```

## 5. Biểu đồ thể hiện tỷ lệ rời bỏ theo dịch vụ điện thoại

```python
# Lọc dữ liệu cho PhoneService và MultipleLines
phone_data = df[df['PhoneService'] != 'No']
multiple_lines_data = df[df['MultipleLines'] == 'Yes']

# Tính toán số lượng và tỷ lệ cho PhoneService
churn_phone = phone_data.groupby(['Churn']).size()
churn_percent_phone = churn_phone / churn_phone.sum() * 100

# Tính toán số lượng và tỷ lệ cho MultipleLines
churn_multiple = multiple_lines_data.groupby(['Churn']).size()
churn_percent_multiple = churn_multiple / churn_multiple.sum() * 100

# Chuẩn bị dữ liệu cho biểu đồ
labels = ['PhoneService', 'MultipleLines']
no_churn = [churn_phone['No'], churn_multiple['No']]
yes_churn = [churn_phone['Yes'], churn_multiple['Yes']]

# Vẽ stacked bar chart
plt.figure(figsize=(10, 8))

# Cột chồng cho "Không rời bỏ"
bar1 = plt.bar(labels, no_churn, label='Không rời bỏ', color=sns.color_palette("Blues", 2)[0])

# Cột chồng cho "Rời bỏ"
bar2 = plt.bar(labels, yes_churn, bottom=no_churn, label='Rời bỏ', color=sns.color_palette("Blues", 2)[1])

# Thêm nhãn số lượng trên đầu các cột
for i in range(len(labels)):
    total_customers = no_churn[i] + yes_churn[i]
    plt.text(i, total_customers + 2, f"{int(total_customers)}", ha='center', fontsize=16, fontweight='bold')

# Thêm nhãn tỷ lệ phần trăm bên trong cột
for i in range(len(labels)):
    # Nhãn cho phần "Không rời bỏ"
    plt.text(
        i, 
        no_churn[i] / 2,  # Vị trí nhãn ở giữa phần "Không rời bỏ"
        f"{churn_percent_phone['No'] if i == 0 else churn_percent_multiple['No']:.1f}%", 
        ha='center', 
        color='black', 
        fontsize=16, 
        fontweight='bold'
    )
    # Nhãn cho phần "Rời bỏ"
    plt.text(
        i, 
        no_churn[i] + yes_churn[i] / 2,  # Vị trí nhãn ở giữa phần "Rời bỏ"
        f"{churn_percent_phone['Yes'] if i == 0 else churn_percent_multiple['Yes']:.1f}%", 
        ha='center', 
        color='black', 
        fontsize=16, 
        fontweight='bold'
    )

# Tùy chỉnh biểu đồ
plt.title("Tỷ lệ Churn theo PhoneService và MultipleLines", fontsize=14)
plt.xlabel("")
plt.ylabel("Số lượng khách hàng")
plt.xticks(fontsize=14)
plt.tight_layout()
plt.show()
```

## 6. Biểu đồ thể hiện tỷ lệ rời bỏ theo dịch vụ Internet

```python
# Danh sách các cột và điều kiện lọc
columns_to_plot = [
    ('InternetService', 'No'), 
    ('OnlineSecurity', 'Yes'),
    ('OnlineBackup', 'Yes'),
    ('DeviceProtection', 'Yes'),
    ('TechSupport', 'Yes'),
    ('StreamingTV', 'Yes'),
    ('StreamingMovies', 'Yes')
]

# Tạo DataFrame chứa tổng số lượng churn và không churn cho từng dịch vụ
data_list = []

for col, filter_value in columns_to_plot:
    if col == 'InternetService':
        filtered_data = df[df[col] != filter_value]
    else:
        filtered_data = df[df[col] == filter_value]
    
    churn_counts = filtered_data['Churn'].value_counts()
    data_list.append({
        'Service': col,
        'No': churn_counts.get('No', 0),
        'Yes': churn_counts.get('Yes', 0),
        'Total': churn_counts.sum()
    })

# Chuyển dữ liệu thành DataFrame và sắp xếp từ lớn đến nhỏ
churn_df = pd.DataFrame(data_list).sort_values(by='Total', ascending=False)

# Vẽ biểu đồ cột chồng
plt.figure(figsize=(14, 8))
bar1 = plt.bar(churn_df['Service'], churn_df['No'], label='Không rời bỏ', color=sns.color_palette("Blues", 2)[0])
bar2 = plt.bar(churn_df['Service'], churn_df['Yes'], bottom=churn_df['No'], label='Rời bỏ', color=sns.color_palette("Blues", 2)[1])

# Thêm nhãn tổng số lượng khách hàng trên đỉnh cột
for i, total in enumerate(churn_df['Total']):
    plt.text(i, total + 2, f"{total}", ha='center', fontsize=16, fontweight='bold')

# Thêm nhãn tỷ lệ phần trăm bên trong cột
for i in range(len(churn_df)):
    no_val = churn_df['No'].iloc[i]
    yes_val = churn_df['Yes'].iloc[i]
    total = churn_df['Total'].iloc[i]
    
    plt.text(i, no_val / 2, f"{no_val/total*100:.1f}%", ha='center', color='black', fontsize=16, fontweight='bold')
    plt.text(i, no_val + yes_val / 2, f"{yes_val/total*100:.1f}%", ha='center', color='black', fontsize=16, fontweight='bold')

# Tùy chỉnh biểu đồ
plt.title("Tỷ lệ Churn theo các dịch vụ", fontsize=18, fontweight='bold')
plt.ylabel("Số lượng khách hàng", fontsize=16)
plt.xticks(fontsize=16, rotation=45)
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

## 7. Biểu đồ thể hiện tỷ lệ rời bỏ theo phương thức thanh toán

```python
# Tính toán số lượng khách hàng theo PaymentMethod và Churn
churn_payment = df.groupby(['PaymentMethod', 'Churn']).size().unstack(fill_value=0)

# Tính tỷ lệ phần trăm
churn_percent_payment = churn_payment.div(churn_payment.sum(axis=1), axis=0) * 100

# Sắp xếp PaymentMethod theo tổng số lượng khách hàng
churn_payment['Total'] = churn_payment['No'] + churn_payment['Yes']
churn_payment = churn_payment.sort_values(by='Total', ascending=False)

# Vẽ biểu đồ cột chồng
plt.figure(figsize=(12, 8))
bar1 = plt.bar(churn_payment.index, churn_payment['No'], label='Không rời bỏ', color=sns.color_palette("Blues", 2)[0])
bar2 = plt.bar(churn_payment.index, churn_payment['Yes'], bottom=churn_payment['No'], label='Rời bỏ', color=sns.color_palette("Blues", 2)[1])

# Thêm nhãn tổng số lượng khách hàng trên đỉnh cột
for i, total in enumerate(churn_payment['Total']):
    plt.text(i, total + 2, f"{total}", ha='center', fontsize=16, fontweight='bold')

# Thêm nhãn tỷ lệ phần trăm bên trong cột
for i, payment in enumerate(churn_payment.index):
    no_val = churn_payment['No'][payment]
    yes_val = churn_payment['Yes'][payment]
    total = no_val + yes_val
    
    # Phần "Không rời bỏ"
    plt.text(i, no_val / 2, f"{no_val/total*100:.1f}%", ha='center', color='black', fontsize=16, fontweight='bold')
    # Phần "Rời bỏ"
    plt.text(i, no_val + yes_val / 2, f"{yes_val/total*100:.1f}%", ha='center', color='black', fontsize=16, fontweight='bold')

# Tùy chỉnh biểu đồ
plt.title("Tỷ lệ Churn theo phương thức thanh toán", fontsize=18, fontweight='bold')
plt.ylabel("Số lượng khách hàng", fontsize=16)
plt.xticks(fontsize=16, rotation=45)
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
